```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def procesar_pagina(soup):

    data = []

    titulos=[]
    autores=[]
    precios_sin_descuento=[]
    precios_con_descuento=[]

    titulo_libros = soup.find_all('dd', class_='title')
    
    for item in titulo_libros:
        titulo=item.a.text.strip()
        titulos.append(titulo)

    autor_libros = soup.find_all('dd', class_='creator')

    for item in autor_libros:
        autor=item.text.strip()
        autores.append(autor)

    strike=soup.select('p.precio strike')

    for item in strike:
        precio=item.text.strip()
        precios_sin_descuento.append(precio)

    strong=soup.select('p.precio strong')

    for item in strong:
        precio=item.text.strip()
        precios_con_descuento.append(precio)

    min_length = min(len(titulos), len(autores), len(precios_sin_descuento), len(precios_con_descuento))

    for i in range (min_length):
        data.append(
            {
                'Titulo':titulos[i],
                'Autores':autores[i],
                'Precio inicial':precios_sin_descuento[i],
                'Precio final':precios_con_descuento[i]
            }
        )

    df=pd.DataFrame(data)

    df.to_csv("/Users/carlo/OneDrive/Escritorio/Curso-python/Actividad_Pagina_WebScraping/librosDescuentoGonvil.csv", index=False)
    df.to_json("/Users/carlo/OneDrive/Escritorio/Curso-python/Actividad_Pagina_WebScraping/librosDescuentoGonvil.json")

    print(df)
```

```python
def obtener_contenido(url):
    response = requests.get(url)
    return response.content
```

```python
def analizar_contenido(html):
    return BeautifulSoup(html, 'html.parser')
```

```python
def manejar_paginacion(url_base, num_paginas):
    for i in range(1, num_paginas+1):
        url=url_base + str(i) + '&cuantos=20&orden=prioridad%2C+fecha_edicion+desc&titulo=C%E1lculo'
        contenido_pagina=obtener_contenido(url)
        soup= analizar_contenido(contenido_pagina)
        procesar_pagina(soup)
```

```python
url_base='https://www.gonvill.com.mx/busqueda/listaLibros.php?pagSel='

num_paginas=10

manejar_paginacion(url_base,num_paginas)
```
